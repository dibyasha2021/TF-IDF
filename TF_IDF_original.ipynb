```python
#MULTIPLE DOCS
Doc1="A spectrogram is a visual depiction of the spectrum of frequencies of an audio signal as it varies with time. "
Doc2="feature extraction is a step in signal processing."
Doc3="An audio signal is a representation of sound waveforms. It encodes all the necessary information required to reproduce sound. Audio signals come in two basic types: analog and digital."
```

```python

#MULTIPE
AllDoc2=[Doc1,Doc2,Doc3]
#AllDoc
```

```python
# After removing the stopword from all the documents
New_alldoc2=[]
for doc in AllDoc2:
  New_alldoc2.append(remove_stopwords(doc))
print(New_alldoc2)
```

```python
# tokenization and lemmetization
def tokenlemma(alldoc):
  lemmatizer = WordNetLemmatizer()
  Documents=[]
  for doc in alldoc:
    sentence=" ".join(re.split(r'[-\s.,;:!?()♁]+',doc))
    tokens=word_tokenize(sentence)
    text=[]
    for token in tokens:
      if token == any('’','s'):
        continue
      text.append(lemmatizer.lemmatize(token))
    Documents.append(text)
  return Documents


#word_list_single=tokenlemma(New_alldoc1)
word_list_multi=tokenlemma(New_alldoc2)
```

```python
#finding Unique words in all  the documents
Unique_word_set=np.union1d(word_list_multi[0],word_list_multi[1])
wordset=np.union1d(Unique_word_set,word_list_multi[2])

print(wordset)
```

```python
# Calculating Bag of words
def calculateBOW(wordset,doc):
  tf_diz = dict.fromkeys(wordset,0)
  for word in doc:
      tf_diz[word]=doc.count(word)
  return tf_diz
```

```python
#BOW of All doc
Bag_of_words=[]
for doc in word_list_multi:
    Bag_of_words.append(calculateBOW(wordset,doc))
#Bag_of_words
```

```python
# Term frequency function
def termFrequency(term, doc):
        
    # Splitting the document into individual terms
    normalizeTermFreq = doc.lower().split()
 
    # Number of times the term occurs in the document
    term_in_document = normalizeTermFreq.count(term.lower())
 
    # Total number of terms in the document
    len_of_document = float(len(normalizeTermFreq ))
 
    # Normalized Term Frequency
    normalized_tf = term_in_document / len_of_document
 
    return normalized_tf
```

```python
#TF of the given word in document I
termFrequency('signal',Doc2)
```

```python
#finding Term frequency of each unique words
def Termfreq_all(unique_words,AllDoc):
  tf_word={}
  
  for token in unique_words:
    tf_list_docwise=[]
    for doc in AllDoc:
      tf_list_docwise.append(termFrequency(token,doc))
    
    tf_word[token]=tf_list_docwise
        
  return tf_word
```

```python
# TF of each word in all the Documents
Termfreq_all(wordset,AllDoc2)
```

```python
# The Inverse Document Frequency -idf
def IDF(corpus, unique_words):
   idf_dict={}
   N=len(corpus)
   for i in unique_words:
     count=0
     for sen in corpus:
       if i in sen.split():
         count=count+1
       idf_dict[i]=(math.log((1+N)/(count+1)))+1
   return idf_dict 
```

```python
#IDF of each for all the documents cumulatively 
idf=IDF(AllDoc2,wordset)
idf
```

```python
#TF-IDF score each term in a given document
tf=termFrequency('signal',Doc2)
idf_term=idf['signal']
Score=tf*idf_term

print(Score)

```

```python
# TF-IDF Score wor all unique words 
def Score(unique_words,AllDoc):
  idf=IDF(AllDoc,unique_words)
  score={}
  for token in unique_words:
    Score=[]
    for doc in AllDoc:
      tf=termFrequency(token,doc)
      Score.append(tf*idf[token])

    score[token]=Score
  return score
final_score=Score(wordset,AllDoc2)
```

```python
# function to get the TF-IDF score value document wise
FS_df=pd.DataFrame.from_dict(final_score,orient="index",columns=['Doc1','Doc2','Doc3'])
FS_df
```

```python
result=FS_df.sort_values(by='Doc1',ascending=False)
result
```
